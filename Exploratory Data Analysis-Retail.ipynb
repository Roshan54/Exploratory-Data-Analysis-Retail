```python
#Importing all the necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading data from the given dataset

data = pd.read_csv("SampleSuperstore.csv")
data.head()
```

```python
# Checking the info of the given data

data.info()
```

```python
# Describing the data in the given dataset

data.describe()
```

Here we can notice that the data does not comprise of any null values.

```python
# Checking for duplicate values in the data

data.duplicated().sum()
```

```python
# Dropping the duplicate values from the given data

data.drop_duplicates(subset=None,keep='first',inplace=True)

# Rechecking for the presence of duplicate values
data.duplicated().sum()
```

```python
# Dropping the columns that are not useful for our analysis

data.drop(['Postal Code'], axis=1)
```

```python
data.head()
```

### Exploratory Data Analysis(EDA)

```python
# Plotting barplots for sales and profit by Product Category

fig, ax = plt.subplots(1,2, figsize=(13,6))

data_category = data.groupby("Category")[["Sales", "Profit", "Region"]].sum()

ax[0].set_title("Sales by Product Category")
sns.barplot(x = data_category.index, y = data_category["Sales" ], ax=ax[0])

ax[1].set_title("Profit by Product Category")
sns.barplot(x = data_category.index, y = data_category["Profit"], ax=ax[1])

fig.show()
```

Analyzing the above bar plot we can conclude that the product category <b>Furniture</b> makes the least profit even though it has a reasonable amount of sales, whereas <b>Technology</b> has the best sales and profit figures out of all the product categories.

```python
# Plotting barplots for Sales and Profit by Region

data_region = data.groupby("Region", sort=True)[["Sales", "Profit"]].sum()

fig, ax = plt.subplots(1 ,2, figsize=(13,6))

ax[0].set_title("Sales by Region")
sns.barplot(x = data_region.index, y = data_region["Sales"], ax = ax[0])
 
ax[1].set_title("Profit by Region")
sns.barplot(x = data_region.index, y = data_region["Profit"], ax = ax[1])

plt.show()
```

The <b>Western</b> region shows maximum Sales and Profit while the <b>Central</b> region has the Lowest Profit among all even though it has the third highest sales.

```python
# Looking at the distribution of products over different regions

plt.figure(figsize = (13, 6))
sns.countplot(x = "Region", hue = "Category", data = data)
```

Here we can notice that among all the Regions, Office Supplies has the Maximum Product Count.

```python
# Visualizing the profits of subcategory by making use of the pie chart

profit = data[data.Profit > 0]

plt.pie(profit.groupby('Sub-Category').agg('sum').Profit, radius=3.12, labels = profit.groupby('Sub-Category').agg('sum').index,
       autopct = '%1.2f%%')
plt.title('Profit Pie Chart',fontdict = dict(fontsize = 20), pad = 100,loc = 'center')
plt.show()
```

Here we can observe that <b>Binder(15.54%)</b> shows the highest Profit followed by <b>Copiers(12.57%)</b> and <b>Phones(11.77%)</b>.

```python

loss = data[data.Profit < 0]
plt.pie(np.abs(loss.groupby('Sub-Category').agg('sum').Profit), radius = 3.12, labels = loss.groupby('Sub-Category').agg('sum').index,
       autopct = '%1.2f%%')
plt.title('Loss Pie Chart',fontdict = dict(fontsize = 20), pad = 100, loc = 'center')
plt.show()
```

Here we can observe that <b>Binder</b>, <b>Tables</b> and <b>Machines</b> have the highest loss. On the other hand <b>Copier</b>, <b>Paper</b> and <b>Labels</b> shows no loss, which could be due to the reason that all of them are Office Accessories.

```python
# Lets find the State with the most number of buyers

plt.figure(figsize=(15,12))
sns.countplot(x = 'State',data = data, palette = 'deep', order = data['State'].value_counts().index)
plt.xticks(rotation = 90)
plt.show()
```

Here we can notice <b>California</b> has the highest number of buyers followed by <b>New York</b> and <b>Texas</b>.

```python
# Lets find the subcategories with the highest/lowest sales and profit

ps = data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['aqua','blue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

From the graph above we can define that <b>Phones</b> and <b>Chairs</b> have the highest sales whereas <b>Fasteners</b> and <b>Labels</b> have the least sales.

In terms of profit, <b>Copiers</b>, <b>Phones</b> and <b>Accessories</b> yield the most profit.

In terms of loss, <b>Tables</b> and <b>Bookcases</b> are the only product subcategories with negative profits
